Importing the Dataset

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

Data Preprocessing

```python
dataset=pd.read_csv(r"..\Datasets\Market_Basket_Optimisation.csv",header= None)
transctions=[]
for i in range ( 0,7501):
    transctions.append([str(dataset.values[i,j])for j in range(0,20)])
```

```python
print(transctions)
```

Training Apriori model on dataset

```python
from apyori import apriori
rules=apriori(transactions=transctions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

visualising the results

```python
results=list(rules)
print(results)
```

Putting the results into well organised pandas dataframe

```python
def inspect (results):
    lhs =[tuple(result[2][0][0])[0]for result in results]
    rhs =[tuple(result[2][0][1])[0]for result in results]
    supports=[result[1]for result in results]
    confidences=[result[2][0][2]for result in results]
    lifts=[result[2][0][3]for result in results]
    return list (zip(lhs,rhs,supports,confidences,lifts))
resultsinDataFrame=pd.DataFrame(inspect(results),columns=['Left Hand Side',"Right Hand Side","Supports",'Confidence',"Lift"])
```

Displaying the results non sorted

Displaying the results sorted by Decending Lifts

```python
resultsinDataFrame.nlargest(n=10,columns="Lift")
```
