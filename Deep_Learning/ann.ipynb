Artificial Neural Networks(ANN)

Importin libraries

```python
import numpy as np 
import pandas as pd 
import tensorflow as tf
```

```python
tf.__version__
```

Part1 :- Data Preprocessing

Importing the dataset

```python
dataset=pd.read_csv(r"D:\python\ML & AI Pratice\Datasets\Churn_Modelling.csv")
X=dataset.iloc[:,3:-1].values
y=dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

Encoding categorical data

Label Encoding the "Gender" column

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,2]=le.fit_transform(X[:,2])
```

One Hot Encoding the "Geography" column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[1])],remainder="passthrough")
X = np.array(ct.fit_transform(X), dtype=np.float32)
```

```python
print(X)
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

Part2 :- Building ANN

Initializing the ANN

```python
ann=tf.keras.models.Sequential()
```

Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

Adding the Output layer

```python
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

Part3 :- Training Ann

Compiling ANN

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

Initalizing the ANN

```python
ann.fit(X_train, y_train, batch_size=32, epochs=25)
```

Training the ANN on the Training set

```python
customer = np.array([[1,0,0,600,1,40,3,60000,2,1,1,50000]])
customer_scaled = sc.transform(customer)
prediction = ann.predict(customer_scaled)
print(prediction)
```

Part4 :-Marking the predictions and evaluating the model

```python
print(ann.predict(sc.transform( np.array([[1,0,0,600,1,40,3,60000,2,1,1,50000]]))))
```

Predicting the test set results

```python
y_pred=ann.predict(X_test)
y_pred=(y_pred>0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

making confusiton matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
