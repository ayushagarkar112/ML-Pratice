```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os 
os.chdir('..')
```

Importing The Dataset

```python
dataset=pd.read_csv(r"D:\python\ML & AI Pratice\Datasets\Position_Salaries.csv")
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

Creating The Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

Polynomial Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)
x_poly=poly_reg.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)
```

Visualizing The Results

```python
plt.scatter(x,y,color="red")
plt.plot(x,lin_reg.predict(x),color="blue")
plt.title('Position Vs Salaries (Linear Regression)')
plt.xlabel("position")
plt.ylabel("salaries")
plt.show
```

Visualizing for Polynomial Linear Regression

```python
plt.scatter(x,y,color="red")
plt.plot(x,lin_reg_2.predict(x_poly),color="blue")
plt.title('Position Vs Salaries (Polynomial Linear Regression)')
plt.xlabel("position")
plt.ylabel("salaries")
plt.show
```

Visulazing the Polynomial Regression (For Smoother Model)

```python
#Retraining the Model For Smoother Result
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly_2=poly_reg.fit_transform(x)
lin_reg_3=LinearRegression()
lin_reg_3.fit(x_poly_2,y)
```

```python
plt.scatter(x, y, color="red")
plt.plot(x, lin_reg_3.predict(x_poly_2), color="blue")
plt.title("Position Vs Salaries (Polynomial Linear Regression)")
plt.xlabel("position")
plt.ylabel("salaries")
plt.show()
```

For Extra Smooth Graph (Using Small Incremental Values)

```python
x_grid = np.arange(x.min(), x.max(), 0.1)
x_grid=x_grid.reshape((len(x_grid),-1))
X_grid_poly = poly_reg.transform(x_grid)

plt.scatter(x, y, color='red')
plt.plot(x_grid, lin_reg_3.predict(poly_reg.fit_transform(x_grid)), color="blue")
plt.title("Position Vs Salaries (Polynomial Linear Regression)")
plt.xlabel("position")
plt.ylabel("salaries")
plt.show()
```
