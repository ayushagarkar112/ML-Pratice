Import the Liblaries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import os 
os.chdir("..")
```

Import the Dataset

```python
dataset=pd.read_csv(r"D:\python\ML & AI Pratice\Datasets\Position_Salaries.csv")
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values
```

Training Random Forest Regression on whole Dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor= RandomForestRegressor(n_estimators=10,random_state=0)
regressor.fit(X,Y)
```

Predicting new Result

```python
regressor.predict([[6.5]])
```

Visualising the result

```python
X_grid = np.arange(X.min(), X.max(), 0.1)
X_grid=X_grid.reshape((len(X_grid),-1))
plt.scatter(X, Y, color='Green')
plt.plot(X_grid, regressor.predict(X_grid), color="Red")
plt.title("Position Vs Salaries (Random Forest  Regression)")
plt.xlabel("position")
plt.ylabel("salaries")
plt.show()
```
