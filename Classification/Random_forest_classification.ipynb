Random Forest Classification

Importing the Liblaires

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

Loading the Dataset

```python
dataset=pd.read_csv(r"D:\python\ML & AI Pratice\Datasets\Social_Network_Ads.csv")
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

Splitting Dataset into Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

Training Random Forest Classifer on Training set

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=10,criterion='entropy',max_depth=4,random_state=0)
model.fit(X_train,y_train)
```

Predicting the Result

```python
print(model.predict(sc.transform([[19,19000]])))
```

Predicting Test Results

```python
y_pred=model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Making Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

Visualising The Result

```python
sns.set(style="whitegrid")

# Combine train and test for boundary plotting
X_all = np.vstack((X_train, X_test))
y_all = np.hstack((y_train, y_test))

# Create mesh grid
x_min, x_max = X_all[:, 0].min() - 1, X_all[:, 0].max() + 1
y_min, y_max = X_all[:, 1].min() - 1, X_all[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

# Predict over grid
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3)

# Plot training points
sns.scatterplot(
    x=X_train[:, 0],
    y=X_train[:, 1],
    hue=y_train,
    palette="deep",
    edgecolor="k",
    s=60,
    legend=True
)

# Plot test points
sns.scatterplot(
    x=X_test[:, 0],
    y=X_test[:, 1],
    hue=y_test,
    palette="deep",
    markers="X",
    s=80,
    legend=False
)

plt.title("Kernel SVM Decision Boundary (Random Forest Classifer)")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.show()
```
