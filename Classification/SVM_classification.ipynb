SVM Classifer

Loading the liblaries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

Loading the dataset

```python
dataset=pd.read_csv(r"..\Datasets\Social_Network_Ads.csv")
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

Splitting the dataset into training and test dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

Running the SVM model on training set

```python
from sklearn.svm import SVC
classifer=SVC(kernel='linear',random_state=0)
classifer.fit(X_train,Y_train)
```

Predicting the Results

```python
result=classifer.predict(sc.transform([[35,20000]]))
print(result)
```

Predicting the test set results

```python
Y_pred=classifer.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1),Y_test.reshape(len(Y_test),1)),1))
```

Making The confusion matrix

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm=confusion_matrix(Y_test,Y_pred)
print(cm)
accuracy_score(Y_test,Y_pred)
```

Visualising the Result

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create DataFrame using YOUR variables
df = pd.DataFrame(X_train, columns=["Age", "Estimated Salary"])
df["Actual"] = Y_train
df["Predicted"] = classifer.predict(sc.transform(X_train))

# Plot using seaborn (LOW MEMORY, SAFE)
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="Age",
    y="Estimated Salary",
    hue="Predicted",
    style="Actual",
    palette=["red", "green"],
    s=60
)

plt.title("KNN Classification (Training set)")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.legend(title="Classes")
plt.show()
```
